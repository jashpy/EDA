```python
df1=pd.read_csv('country-code.csv')
df1.head()
```

```python
df1.rename(columns = {'country_code':'Country Code'}, inplace = True)
df1.rename(columns = {'country':'Country'}, inplace = True)
```

```python
df1.head()
```

```python
df2=pd.merge(df,df1,on='Country Code', how='left')
```

```python
df2.head()
```

```python
df2.columns
```

```python
ax=sns.countplot(x=df2["Country"])
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
```

**Top three countries for transactions:**

**1.India**

**2.United States**

**3.United kingdom**

```python
ratings=df2.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings.head()
```

```python
sns.scatterplot(x='Rating text',y='Aggregate rating',data=ratings)
```

```python
plt.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

**Not Rated count is very high**

**Maximum number of rating are between 2.5 to 3.4**

```python
sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

**Most people has rated it as average**
