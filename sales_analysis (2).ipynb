```python
import pandas as pd
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt
import seaborn as sns
df=pd.read_csv('sales.csv')

df.head(5)
```

```python
df.info()
```

```python
pd.isnull(df).sum()                      #checking for null values
```

```python
df['Amount']=df['Amount'].astype('int')                 #changing data types
```

```python
df['Amount'].dtypes
```

Gender

```python
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)                      #from these we can visualize women customers are more in numbers
```

```python
df1=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
df1
```

```python
sns.barplot(x='Gender',y='Amount',data=df1)
```

```python
#gender conclusion:
#female purchase more and have more purchasing power
```

AGE

```python
df.columns
```

```python
ax=sns.countplot(x='Age Group',hue='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df2=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
df2
```

```python
sns.barplot(x='Age Group',y='Amount',data=df2)
```

```python
#people between 26-35 have more purachasing power and here also females wins
```

STATE

```python
ax=sns.countplot(x='State',data=df)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df3=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
df3
```

```python
ax=sns.barplot(x='State',y='Amount',data=df3)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
```

```python
#from here we can conclude people from uttar pradesh have the maximum buyers while minimum in punjab
```

MARITAL STATUS

```python
ax=sns.countplot(x='Marital_Status',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df4=df.groupby(['Marital_Status'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
df4
```

OCCUPATION

```python
ax=sns.countplot(x='Occupation',data=df)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df5=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
df5
```
